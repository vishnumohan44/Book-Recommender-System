```python
import pandas as pd
import numpy as np
```

Reading all the files

```python
books = pd.read_csv('data/BX-Books.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
books.columns
```

```python
features = ['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']
books = books[features]
```

```python
users = pd.read_csv('data/BX-Users.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
users.rename(columns={'User-ID':'user','Location':'location','Age':'age'},inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
ratings.rename(columns={'User-ID':'user','Book-Rating':'rating'},inplace=True)
```

```python
ratings.head()
```

Data Preprocesing

```python
users.shape
```

```python
books.shape
```

```python
ratings.shape
```

```python
ratings['user'].value_counts().shape
```

```python
x = ratings['user'].value_counts()>200
x[x]
```

```python
y = x[x].index
y
```

```python
ratings = ratings[ratings['user'].isin(y)]
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books.shape
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('book')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'no_of_ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating, on='book')
final_rating
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['no_of_ratings']>=50]
final_rating.shape
```

```python
final_rating.drop_duplicates(['user','book'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user',index='book',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

Training the Model

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions = model.kneighbors(book_pivot.iloc[54,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestions[0])):
        if i==0: print("Suggestions for books similar to",book_name,'are:')
        else: print(i,book_pivot.index[suggestions[0][i]])
```
